```
#Import libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from pandas.plotting import lag_plot
from pandas import datetime
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
import pandas_datareader as web
```

**Get stock data and split into train-test datsets**

```
#Get the stock quote 
df = web.DataReader('AMD', data_source='yahoo', start='2016-07-22', end='2020-10-30')
#df = web.DataReader('DIS', data_source='yahoo', start='2016-07-22', end='2020-05-26') 
#Show the data 
df.head(5)
```

```
#Train-test split
train_data, test_data = df[0:int(len(df)*0.8)], df[int(len(df)*0.8):]
train_ar = train_data['Close'].values
test_ar = test_data['Close'].values
```

**Build model and predict on testing data**

```
def smape_kun(y_true, y_pred):
    return np.mean((np.abs(y_pred - y_true) * 200/ (np.abs(y_pred) + np.abs(y_true))))
```

```
#Train and test model
train_ar = train_data['Close'].values
test_ar = test_data['Close'].values
history = [x for x in train_ar]
print(type(history))
predictions = list()
for t in range(len(test_ar)):
    model = ARIMA(history, order=(5,1,0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    #print(yhat)
    predictions.append(yhat)
    obs = test_ar[t]
    #history.append(obs)
    history.append(yhat[0])
error = mean_squared_error(test_ar, predictions)
print('Testing Mean Squared Error: %.3f' % error)
error2 = smape_kun(test_ar, predictions)
print('Symmetric mean absolute percentage error: %.3f' % error2)
```

**Preprocess data and split into train-test datasets**

```
#Create a new dataframe with only the 'Close' column
data = df.filter(['Close'])
#Converting the dataframe to a numpy array
dataset = data.values
#Get /Compute the number of rows to train the model on
training_data_len = math.ceil( len(dataset) *.8)
```

```
#Perform scaling
scaler = MinMaxScaler(feature_range=(0, 1)) 
scaled_data = scaler.fit_transform(dataset)
```

**Build RNN model and train on training data**

```
#Build model
def build_model(layers):
    model = Sequential()

    # By setting return_sequences to True we are able to stack another LSTM layer
    model.add(GRU(units=512, return_sequences=True,input_shape=(x_train.shape[1],1)))
    model.add(Dropout(0.2))

    model.add(GRU(units=256))
    model.add(Dropout(0.2))
    model.add(Dense(units=1))
    model.add(Activation("linear"))

    start = time.time()
    model.compile(loss="mse", optimizer="rmsprop", metrics=['accuracy'])
    print("Compilation Time : ", time.time() - start)
    return model
```

```
print("X_train", x_train.shape)
print("y_train", y_train.shape)
print("X_test", x_test.shape)
print("y_test", y_test.shape)
```

```
window = 60
model = build_model([1, window, 100, 1])
```

```
model.fit(
    x_train,
    y_train,
    batch_size=1,
    epochs=1)
```

<font color="green">**Drift detection using weighted ensemble**</font>

**Detect drift and visualize drifts**

```
#mean absolute error
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(y_test, weightedFinalPred)
mae
```

```
#Find drift data
df1=df.reset_index()
datelist=[]
for i in range(len(y_test)):
  diff=abs(y_test[i]-weightedFinalPred[i])
  if(diff>(mae+2.5)):
    #print(df1.loc[training_data_len+i,"Date"])
    datelist.append(df1.loc[training_data_len+i,"Date"])
```

```
#Plot data on graph
train = data[:training_data_len]
valid = data[training_data_len:]
valid['Predictions'] = weightedFinalPred
#Visualize the closing price history
plt.figure(figsize=(16,8))
plt.title('Sudden Drift Detection using Weighted Ensemble: Tesla')
plt.plot(df1['Date'],df1['Close'])
plt.plot(valid[['Close', 'Predictions']])
for i in range(len(datelist)):
  plt.axvline(pd.Timestamp(datelist[i]),color='gray',linestyle='--')
plt.xlabel('Date',fontsize=18)
plt.ylabel('Close Price USD ($)',fontsize=18)
plt.show()
```
